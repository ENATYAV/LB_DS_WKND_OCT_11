```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Functional method

```python
arr1= np.arange(1,10)
```

```python
arr1
```

```python
arr2= arr1**2
arr2
```

```python
plt.plot(arr1,arr2,'r-.')
plt.xlabel('Arr1')
plt.ylabel('Arr1')
plt.title('Arr1 vs Arr2')
```

```python
plt.subplot(2,2,1)
plt.plot(arr1,arr2, 'r-.')
plt.xlabel('Arr1')
plt.ylabel('Arr1')
plt.title('Arr1 vs Arr2')
plt.subplot(2,2,2)
plt.plot(arr2,arr1, 'g-.')
plt.xlabel('Arr1')
plt.ylabel('Arr1')
plt.title('Arr2 vs Arr1')
plt.subplot(2,2,3)
plt.plot(arr1,arr2, 'b-.')
plt.xlabel('Arr1')
plt.ylabel('Arr1')
plt.title('Arr1 vs Arr2')
plt.subplot(2,2,4)
plt.plot(arr2,arr1, 'b-.')
plt.xlabel('Arr1')
plt.ylabel('Arr1')
plt.title('Arr1 vs Arr2')
plt.tight_layout()
```

# Object oriented Method

```python
fig=plt.figure()
ax= fig.add_axes([0.1,0.1,0.9,0.9]) # # [Left Bottom, width Height]
ax.plot(arr1,arr2,'r-.')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

### inset graphs--> Graphs inside a graph 

```python
fig= plt.figure()
ax1= fig.add_axes([0.1,0.1,0.8,0.8])  # main axes
ax2= fig.add_axes([0.2,0.5,0.4,0.3])  # inset axes
ax1.plot(arr1,arr2,'r-.')
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')

ax2.plot(arr2,arr1,'g-.')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')
```

### subplots()--not required to set the axes manually, it will be takig care of it automatically

```python
fig,ax = plt.subplots()  # single graph
ax.plot(arr1,arr2,'r--')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr1')
ax.set_title('Arr1 vs Arr2')
```

```python
fig,ax = plt.subplots(nrows=3,ncols=3,figsize=(12,8),dpi=100)  # for multiple graphs, figsize- to set figure size, dpi---> to increase pixels
plt.tight_layout() # to arrange layout
```

```python
ax # shows all graphs location in list
```

```python
fig,ax = plt.subplots(nrows=3,ncols=3,figsize=(12,8),dpi=100)
ax[0][0].plot(arr1,arr2,'r-')
ax[0][1].plot(arr1,arr2,'g-')
ax[0][2].plot(arr1,arr2,'b-')
ax[1][0].plot(arr1,arr2,'r-')
ax[1][1].plot(arr1,arr2,'g-')
ax[1][2].plot(arr1,arr2,'b-')
ax[2][0].plot(arr1,arr2,'r-')
ax[2][1].plot(arr1,arr2,'g-')
ax[2][2].plot(arr1,arr2,'b-')
plt.tight_layout()
```

## legend vs label

```python
fig, axes = plt.subplots()
axes.plot(arr1, arr2, 'r', label = 'Arr1 vs Arr2')
axes.plot(arr2, arr1, 'g', label = 'Arr2 vs Arr1')
axes.legend(loc = 0)  # 0 is best location, 1->top rigt, 2 top left, 3- bottm left,4->bottom right so on
```

# Matplotlib - color, linewidths and linetypes

```python
arr1 = np.arange(1,10)
arr1
```

```python
arr2 = arr1**2
arr2
```

```python
fig, ax = plt.subplots()
ax.plot(arr1, arr2, 'b--')
ax.plot(arr2, arr1, 'r.-')
```

```python
# line colour, line width & line style
```

```python
fig, ax = plt.subplots()
ax.plot(arr1, arr2, color = 'red', lw = 3, ls = '-')
ax.plot(arr2, arr1, color = 'blue', lw = 10, ls = '--')
```

```python
#marker , makersize, marker face colour,marker edge width & markeredgecolor
```

```python
fig, ax = plt.subplots()
ax.plot(arr1, arr2,'r-',marker='v')
ax.plot(arr2, arr1, 'b-.',marker='v',markersize = 8, markerfacecolor = 'yellow',
       markeredgewidth = 1, markeredgecolor = 'red')
```
