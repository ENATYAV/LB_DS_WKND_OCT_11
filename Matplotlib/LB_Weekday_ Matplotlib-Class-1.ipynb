# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr1 = np.arange(1,10)
arr1
```

```python
arr2 = arr1**2
arr2
```

```python
plt.plot(arr1, arr2, 'r--', )
plt.xlabel('ARR1')
plt.ylabel('ARR2')
plt.title('ARR1 vs ARR2')
plt.show()
```

# Object Oriented Method

```python
# Empty canvas
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])  # [Left Bottom, width Height]

axes.plot(arr1, arr2, 'r')
axes.set_xlabel('ARR1')
axes.set_ylabel('ARR2')
axes.set_title('ARR1 vs ARR2')


```

```python
# Inset Graph - graph inside a graph
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

axes1.plot(arr1, arr2, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

axes2.plot(arr2, arr1, 'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

# Subplots() - not required to set the axes manually, it will be takig care of it automatically

```python
fig,axes = plt.subplots()

axes.plot(arr1, arr2, 'r')
axes.set_xlabel('ARR1')
axes.set_ylabel('ARR2')
axes.set_title('ARR1 vs ARR2')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3, figsize = (12,8), dpi = 100)
plt.tight_layout()
```
