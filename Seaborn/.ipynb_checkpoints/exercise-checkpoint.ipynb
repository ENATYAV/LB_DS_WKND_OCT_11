## rug plot --> shows all datapoints

```python
## second way 

sns.rugplot(pokemon_data['Attack Point'],height=0.5, axis='x') 
# height can be set b/w 0-1 default=0.5 
# axis can set 'y', default is axis='x'
```

```python
my_plot=sns.distplot(pokemon_data['Attack Point'],kde=False,rug=True,bins=100) #bins value can be any thing
my_plot.set_title('Attack Point Distribution')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

# 2.Bivariant Analysis : Scatter plot,hexbin,kde
##  by jointplot()
##  Analysis done by two variables

## 1. Scatter plot

```python
sns.jointplot(x='Attack Point',y='Defense Point', data=pokemon_data)
plt.show()
```

### setting x limit and y limit

```python
sns.jointplot(x='Attack Point',
              y='Defense Point', 
              data=pokemon_data,
              xlim=(0,450),
              ylim=(0,200))
plt.show()
```

## 2. hexbin plot

```python
sns.jointplot(x='Attack Point',
              y='Defense Point', 
              data=pokemon_data,
              kind='hexbin',
              color='b')
plt.show()
```

## 3. kde plot

```python
sns.jointplot(x='Attack Point',
              y='Defense Point', 
              data=pokemon_data,
              kind='kde',
              color='b')
plt.show()
```

## 4. regression plot

```python
sns.jointplot(x="Attack Point",
              y="Defense Point",
              data=pokemon_data,
              xlim={0,160},
              ylim={0,160},
              kind='reg')
```

# 3.correlation Analysis

```python
cormat=pokemon_data.corr()
plt.subplots(figsize=(7,7))
sns.heatmap(cormat,vmax=.8,
            square=True,
            annot=True,
            fmt='.2f',
            cmap= 'Blues') # color can be any like 'Reds'
plt.show()
#vmax--> for color desinty
# annot--> shows values in squares
#fmt--> values till 2 decimal points
#cmpa--> color map
```

# 3.visualization lenear relationship

```python
sns.lmplot(x='Speed Point',
          y='Defense Point',
          data=pokemon_data)
plt.show()
# blue line is regression line/convergension line/line of best fit
# blue shade is confidence interval
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           x_estimator = np.mean)
plt.show()
# x_estimator can be any stastical valve, linke std,median
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
          col='Type1')
plt.show()
# Type1 is column in pokemone_data which have categorical data
```

# 4. Categorical Plot
##   using stripplot(), swarm plot,violin plot, box plot, factor plot

```python
tips = sns.load_dataset('tips') # tips data set is in seaborn liberary which is loaded
print(tips.head())
```

## 1.strip plot

```python
sns.stripplot(x='day',
             y='tip',
             data=tips)
plt.show()
# either x or y shold be categorical data
# here x is categorical data
```

```python
plt.subplots(figsize=(6,6))
sns.stripplot(x='day',
             y='tip',
             data=tips)
plt.show()
```

```python
# jitter -->will decrase the coincidence 
plt.subplots(figsize=(15,5))
sns.stripplot(x='day',
             y='tip',
             data=tips,
             jitter=True)
plt.show()
```

## 2.Swarm plot

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
             y='tip',
             data=tips)
             
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips, 
             hue = 'size')
plt.show()
```

## 3. BOX plot /vicsor plot`

```python
plt.subplots(figsize=(15,5))
sns.boxplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

## 4. violin plot-->combinition of box and kde plot

```python
plt.subplots(figsize=(15,5))
sns.violinplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

## 5. factor plot

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               hue='size',
               size=5,
               aspect=2,
              kind='bar')
plt.show()
```
