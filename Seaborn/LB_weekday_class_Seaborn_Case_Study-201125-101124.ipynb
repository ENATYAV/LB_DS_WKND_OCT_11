# Matplotlib - color, linewidths and linetypes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1 = np.arange(1,10)
arr1
```

```python
arr2 = arr1**2
arr2
```

```python
fig, ax = plt.subplots()
ax.plot(arr1, arr2, 'b--')
ax.plot(arr2, arr1, 'r.-')
```

```python
fig, ax = plt.subplots()
ax.plot(arr1, arr2, color = 'red', lw = 3, ls = '-', marker = 'v',)
ax.plot(arr2, arr1, color = 'green', lw = 3, ls = '-.', marker = 'v', markersize = 8, markerfacecolor = 'yellow',
       markeredgewidth = 1, markeredgecolor = 'red')
```

# Seaborn

```python
import seaborn as sns
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv')
pokemon_data
```

```python
sns.distplot(pokemon_data['Attack Point'])
```

```python
my_plot = sns.distplot(pokemon_data['Attack Point'], hist = False, rug = True)
my_plot.set_title('Attack point Distribution Plot')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

```python
sns.rugplot(pokemon_data['Attack Point'],
           height = 0.5,
           axis = 'x')
```

```python
my_plot = sns.distplot(pokemon_data['Attack Point'], kde = False, rug = True, bins = 10)
my_plot.set_title('Attack point Distribution Plot')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

```python
my_plot = sns.distplot(pokemon_data['Attack Point'], kde = False, rug = True, bins = 100)
my_plot.set_title('Attack point Distribution Plot')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

```python
sns.set(color_codes = True)
sns.kdeplot(pokemon_data['Attack Point'],
            shade = True,
            color = 'g')
plt.show()
```

# Bivariate analysis

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data)
plt.show()
```

```python
sns.jointplot(x = 'Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              xlim = {0,450},
              ylim = {0,200})
plt.show()
```

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              kind ='hex',
              color ='r')
```

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              kind ='kde',
              color ='r')
```

```python
cormat = pokemon_data.corr()
plt.subplots(figsize = (7,7))
sns.heatmap(cormat, vmax = .8,
            square = True,
            annot = True, 
            fmt ='.2f', 
            cmap = 'Reds')
plt.show()
```

# Linear Relationship

```python
sns.lmplot( x = 'Speed Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

```python
# If the thickness of the strip is smaller around the line then confidence interval is lower else it is higher.
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           x_estimator = np.mean)
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
          col = 'Type 1')
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
          col = 'Type 1',
          hue = 'Type 2')
plt.show()
```

```python
sns.regplot(x =pokemon_data['Attack Point'],
            y =pokemon_data['Defense Point'],
            )
plt.show()
```

```python
sns.jointplot(x="Attack Point",
              y="Defense Point",
              data=pokemon_data,
              xlim={0,160},
              ylim={0,160},
              kind='reg')
```

# Categorical Plot

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.stripplot(x='day',
              y='tip',
              data = tips,
              )
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.stripplot(x='day',
              y='tip',
              data=tips,
              jitter = True
              )
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips)
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips, 
             hue = 'size')
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.boxplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.violinplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               hue='size',
               size=5,
               aspect=2,
              kind='bar')
plt.show()
```
