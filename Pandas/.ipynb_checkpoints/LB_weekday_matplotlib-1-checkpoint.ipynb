```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr1 = np.arange(1,10)
arr1
```

```python
arr2 = arr1**2
arr2
```

# Functional plot

```python
plt.plot(arr1,arr2,'r') # can be 'r,g,b any colour' with  -,-., .,..
plt.xlabel('Arr1')
plt.ylabel('Arr2')
plt.title('Arr1 vs Arr2')
```

```python
# subplot() in functional plot
```

```python
plt.subplot(2,2,1)
plt.plot(arr1, arr2, 'r-.')
plt.subplot(2,2,2)
plt.plot(arr1, arr2, 'g--')
plt.subplot(2,2,3)
plt.plot(arr2, arr1, 'g--')
plt.subplot(2,2,4)
plt.plot(arr2, arr1, 'r-.')
```

# object oriented Method

```python
fig=plt.figure()  # to select canvas
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])  # [Left Bottom, width Height] -->to select area
axes.plot(arr1,arr2, 'r-.')
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr2')
axes.set_title('Arr1 vs Arr2')
```

```python
## inset graph--->graph inside a graph
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) # for first graph axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # for second graph axis
axes1.plot(arr1, arr2, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

axes2.plot(arr2, arr1, 'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

## Subplots() - not required to set the axes manually, it will be takig care of it automatically

```python
fig,axes = plt.subplots()

axes.plot(arr1, arr2, 'r')
axes.set_xlabel('ARR1')
axes.set_ylabel('ARR2')
axes.set_title('ARR1 vs ARR2')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3, figsize = (12,8), dpi = 100)
plt.tight_layout()
```

```python
fig, axes = plt.subplots()
axes.plot(arr1, arr2, 'r', label = 'Arr1 vs Arr2')
axes.plot(arr2, arr1, 'g', label = 'Arr2 vs Arr1')
axes.legend(loc = 0) # 0 is for best fit, 1->right top corner, 2-->let top corner,3-->left bottom, 4-->right  so on
```
